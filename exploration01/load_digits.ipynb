```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
digits_data = load_digits()

digits_df = pd.DataFrame(data=digits_data.data, columns=digits_data.feature_names)
digits_df
```

```python
digits_data.keys()
```

```python
print(digits_data.DESCR)
```

```python
digits_data.target
```

```python
digits_data.target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data.data, 
                                                    digits_data.target, 
                                                    test_size=0.2, 
                                                    random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier 

decision_tree = DecisionTreeClassifier(random_state=10)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=10)
random_forest.fit(X_train, y_train)
y_pred_rf = random_forest.predict(X_test)
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn import svm
svm_cl = svm.SVC()
svm_cl.fit(X_train, y_train)
y_pred_svm = svm_cl.predict(X_test)

print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.linear_model import SGDClassifier
sgd_cl = SGDClassifier()
sgd_cl.fit(X_train, y_train)
y_pred_sgd = sgd_cl.predict(X_test)

print(classification_report(y_test, y_pred_sgd))
```

```python
from sklearn.linear_model import LogisticRegression
logistic_cl = LogisticRegression(max_iter=2499)
logistic_cl.fit(X_train, y_train)
y_pred_logi = logistic_cl.predict(X_test)

print(classification_report(y_test, y_pred_logi))
```

손글씨 분류에서는 svm 모델이 가장 높은 성능을 나타내는 것으로 보인다.
svm의 특성상 다중 클래스인 경우에도 다른 클래스들과 구분되도록 이진 분류 모델을 학습 시키기 때문인것 같다.
